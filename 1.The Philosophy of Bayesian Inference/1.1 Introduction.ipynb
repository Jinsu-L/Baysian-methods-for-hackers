# 1장 베이지안 추론의 철학

## 1.1 서론
### 1.1.1 베이지안 심리 상태
  
- 베이지안 추론은 불확실성을 유지한다는 점에서 전통적인 통계적 추론과 다름
- 베이지안 세계관은 확률을 사건에서 믿을 수 있는 정도를 계량한 척도  
-> 즉, 사건의 발생을 얼마나 자신하는가로 해석

#### 빈도주의 vs 베이지안
  
* 빈도주의
     * 사건이 장기적으로 일어나는 빈도
     * 항공기의 사고 확률 -> 오랫동안 발생한 비행기 사고의 빈도
    
    
* 베이지안
    * 사건발생에 대한 믿음 또는 확신의 척도
    * 확률은 의견이나 견해를 요약한 것
    
#### 베이지안

믿음을 만들기 위한 증거

* 사전 확률 -> 사건 A에 대한 우리의 믿음의 양 : P(A)
* 사후 확률 -> 업데이트된 믿음, P(A|X)는 증거 X가 증거로 주어진 상황에서 A의 확률

예시)
1. 동전 던지기
    * P(A) : 동전의 앞면이 나올 확률
    * P(A|X) : 동전이 앞면으로 떨어지는 걸 본 후 동전의 앞면이 나올 확률,   
    -> 만약 앞면인 것 봤으면 1.0, 뒷면이 나올 확률 은 0.0
    
    
2. 코드의 버그 유무
    * P(A) : 버그가 있을 것 같다.
    * P(A|X) : 코드가 X 테스트를 통과한 후 버그가 있을 정도


3. 환자가 걸린 질병
    * P(A) : 환자가 질병에 걸릴 확률
    * P(A|X) : 혈액검사 X 후 질병에 걸릴 확률
    
    
### 1.1.2 실제 베이지안 추론
* 빈도주의
    * "내 코드가 테스트를 모두 통과했다. 내 코드에는 버그가 없는가?" -> Yes
    * 빈도주의의 추론 함수는 숫자를 반환하여 추정
    
    
* 베이지안
    * "내 코드에는 종종 버그가 있다. 내 코드가 테스트를 모두 통과했다. 내 코드에는 버그가 없는가?"  
    -> 80% Yes, 20% No
    * 확률을 반환
    * 사전정보("종종 버그가 있다")를 포함 시킴으로써 상황에 대한 믿음을 반영
    
    
#### 증거 포함하기
* 더 많은 증거를 얻을 수록, 우리의 사전 믿음은 새로운 증거에 의해 '희석된다'  
* 증거의 개수 N, N -> ∞ 이면, 베이지안의 결과는 (종종) 빈도주의자의 결과와 일치한다.
    * N 이 커질수록 통계적 추론은 어느정도 객관적 <-> 작아질수록 추론은 불안정 
* 베이지안 분석에서는 사전확률을 도입하고 확률을 돌려줌으로써 불확실성을 유지  
<-> 빈도주의자의 추정은 N이 작아질 수록 변동성이 크고, 신뢰구간이 더 넓어진다.


### 1.1.3 빈도 주의자의 방법이 틀렸나?
* No
    * 빈도주의는 여전히 유요, 많은 분야에서 사용됨  
        -> 최소자승선형회귀, 라소회귀, EM 알고리즘
        

### 1.1.4 빅데이터에 대한 논의
* 빅데이터의 예측 분석 문제는 상대적으로 간단한 알고리즘으로 해결된다.
* 빅데이터 예측의 어려움은 사용한 알고리즘 보다는 빅데이터를 저장하고 실행하는데 있다.
* 더 어려운 분석 문제는 미디엄 데이터, 스몰 데이터 이다.
